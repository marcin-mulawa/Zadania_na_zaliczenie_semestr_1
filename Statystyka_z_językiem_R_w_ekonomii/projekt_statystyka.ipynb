###### Obliczenie korelacji pomiędzy cenami a minmalnym wynagrodzeniem.

```r
wynagrodzenie_min_zl <- c(899.1,936,1126,1276,1317,1386,1500,1600,1680,1750,1850,2000,2100,2250)
wynagrodzenie_min_zl
```

```r
towary_rocznie_polska<- c()
j<-1
for(i in 1:167){
    if(i %% 12 == 1){
        year_sum <-0
    }
    year_sum <- year_sum + towary[i,"POLSKA"]
    if(i %% 12 == 0){
        year_mean<-year_sum/12
        towary_rocznie_polska[j] <- year_mean
        j<- j+1
    }
    if(i == 167){
        year_mean<-year_sum/11
        towary_rocznie_polska[j] <- year_mean
        j<- j+1
    }
}
towary_rocznie_polska <- round(towary_rocznie_polska,2)
towary_rocznie_polska
```

```r
cor(towary_rocznie_polska,wynagrodzenie_min_zl)
```

###### Korelcja pomiędzy cenami, a minimalnym wynagrodzeniem jest bardzo silna, z tego wynika, że wysokość minimalnego wynagrodzenia jest silnie powiązana z cenami produktów.

```r
years <- 2006:2019
y <- wynagrodzenie_min_zl/towary_rocznie_polska
plot(years,y,type='l',ylab='Liczba towarów, które Polacy mogą kupić (pod uwagę brana jest średnia cena wszystkich towarów)',
     main='Siła nabywcza Polaków z minimalnym wynagrodzeniem w latach 2006-2019.')
```

###### Z powyższego wykresu można wywnioskować, że siła nabywcza Polaków z minimalnym wynagrodzeniem na przestrzeni lat 2006-2019 znacznie się poprawiła.

#### Własne wnioski.
######  W Polsce osoby zarabiające minimalne wynagrodzenie zarabiają nie tylko więcej na umowie, ale również stać ich na więcej pomimo wzrostu cen. Życie w Polsce niemal w całym przedziale czasowym 2006-2019 najtańsze było w województwie Lubelskim. Nie uzwględniłem jednak wystarczającej ilości danych takich jak chociażby średnie wynagrodzenie w Polsce aby móc wnioskować na temat ogólnego wzrostu lub spdaku dobrobytu w Polsce.
